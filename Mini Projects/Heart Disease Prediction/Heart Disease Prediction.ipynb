# Splitting the data

```python
X = df.drop(columns = 'target')
Y = df["target"]
```

```python
print(X)
```

```python
print(Y)
```

# Train Data & Test Data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,stratify = Y,random_state = 2,test_size =  0.2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Standardization of data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_Standardized = scaler.transform(X_train)
```

```python
print(X_train_Standardized)
```

# Model Training

0 -> no heart disease
1 -> heart disease

```python
model = LogisticRegression(solver='lbfgs', max_iter=200)
```

```python
model.fit(X_train,Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print("The accuracy of the model on training data is : ",training_data_accuracy*100)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print("The accuracy of the model on test data is : ",test_data_accuracy*100,"%")
```

```python
def predictmodel(input_data):
    # converting input data to numpy array
    input_data_as_np_array = np.asarray(input_data)

     # reshape the input numpy array as we are predicting for one instance
    input_data_reshaped = input_data_as_np_array.reshape(1,-1)

    # standardizing the data
    std_data = scaler.transform(input_data_reshaped) #(we do not fit this data because the test data only 20% and do not represent true nature of our data)

    prediction = model.predict(std_data)

    if prediction[0] == 0:
        print("No heart disease")
    else:
        print("Heart disease")

    return prediction
```
